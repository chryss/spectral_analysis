```python
import geopandas as gp
import pandas as pd
from pathlib import Path
```

```python
datapath = Path("/Volumes/FIREICE/fiboreal/cwaigl/2021-HySpex_fuels/")
hyspexpath = Path("/Volumes/FIREICE/fihyper/cwaigl/")
```

Get 20m polygons

```python
hexagons = gp.read_file(
    datapath / "Hex/hexgrid_20_BC.gpkg")
hexagons
```

Get veg points for BC

```python
bc_vegpoints = gp.read_file(
    datapath / "GIS/veg_locations/CW_BC_points.gpkg")
bc_vegpoints
```

Get all BC flighline shapes

```python
collectionevents = ["20200830_BC", "20210803_BC"] 
flightlinesubdir = "02_intermediate/vectormasks_cropped"
```

```python
gdflist = []
for prefix in collectionevents: 
    for pth in sorted((hyspexpath / prefix / flightlinesubdir ).glob("*buff5.shp")):
        print(prefix, pth.name[10:12])
        flightline = gp.read_file(pth)
        selected = gp.sjoin(bc_vegpoints, flightline, how='inner', predicate='within')
        selected['prefix'] = prefix
        selected['flightline'] = pth.name[10:12]
        gdflist.append(selected)
labeledpts = gp.GeoDataFrame(pd.concat(gdflist, ignore_index=True), crs=gdflist[0].crs)
labeledpts.drop(columns=['Id', 'FID', 'index_right'], inplace=True)
```

```python
labeledpts
```

```python
labeled_hex = gp.sjoin(hexagons, labeledpts, how='inner', predicate='contains')
labeled_hex
```

```python
outdir = Path("/Volumes/FIREICE/fiboreal/cwaigl/2021-HySpex_fuels/Field_Data/") 
labeled_hex.to_file(outdir / "BC_testpoint_hex20.gpkg") 
```
